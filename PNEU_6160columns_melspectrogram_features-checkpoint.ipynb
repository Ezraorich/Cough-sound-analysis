```python
def get_data_files():
    # we will be leaving some files out, on purpose. Do not change this function!
    return [os.path.join(DATA_DIR, x) for x in sorted(os.listdir(DATA_DIR)) ]
```

```python
SAMPLE_RATE = 16000
a = []
data = []
myDict = {}
for i in get_data_files():
    #a = load_audio_file(i)
    y, sr = librosa.load(i)
   
    val = []
    key = get_sound_id_from_file_path(i)
        #ZCR = zero_crossing_rate(y, FRAME_LENGTH, HOP_LENGTH)
    data = librosa.feature.melspectrogram(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    S = np.abs(librosa.stft(y))
    contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
    flatness = librosa.feature.spectral_flatness(y=y)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr, roll_percent=0.99)
    rms = librosa.feature.rms(y=y)



        
        
# #
        #val.extend(zcr)
    val.extend(data)
    val.extend(spec_bw)
    val.extend(cent)
    val.extend(contrast)
    val.extend(flatness)
    val.extend(rolloff)
    val.extend(rms)
    
       
# #
        
        
        
    myDict[key] = val
# #         # print( dur )
# #
pneu = pd.DataFrame.from_dict(myDict, orient='index')
```
